```python
#!/usr/bin/env python
# coding: utf-8

# In[1]:


import spacy;
import nltk;
from nltk.tokenize import sent_tokenize;
from nltk import pos_tag
from nltk import RegexpParser
nlp=spacy.load('en_core_web_sm');
from nltk.corpus import wordnet 
from nltk.tokenize.regexp import regexp_tokenize;
import re
import requests;
import json;
import random;


# In[2]:


import gensim, multiprocessing;
from gensim.models import Word2Vec, KeyedVectors;
import pandas as pd;


# In[3]:


text = '''A computer is a machine that can be instructed to carry out sequences of arithmetic or logical operations automatically via computer programming. Modern computers have the ability to follow generalized sets of operations, called programs.'''


# In[4]:


# Check if any noun or superlatives is present in the sentence...

def check1(key_list):
    
    for i,j in key_list:
        if j=="NN" or j=="NNS" or j=="NNP" or j=="JJS":
            return True;
        
    return False;    
    


# In[5]:


# Check if any entity is present in the sentence...

def check2(entity_set):
    
    if entity_set.ents:
        return True;
    else:
        return False;
    


# In[6]:


# Adding all entities present in the database to our databse...

def add(entity_set):
    
    for ent in entity_set.ents:
        alternative_choice.update({ent.text: ent.label_});


# In[7]:


# Adding all tags to our potential key-list...

def get_key(key_list):
    
    for i,j in key_list:
        if j=="NN" or j=="NNS" or j=="NNP": #or j=="JJS":
            potential_key[i]=j;
            best_list[i]=text.count(i);


# In[8]:


def get_alt(blank):
    
    df = pd.read_csv("RedditWorldNews.csv");

    newsTitle = df['title'].values;
    newsVec = [nltk.word_tokenize(title) for title in newsTitle]
    model = Word2Vec(newsVec, min_count=1, size=32) 

       
    final = model.wv.most_similar(blank);
    r1 = random.randint(0, 3); 
    choices=[];
    
    choices.append(final[0][0]);
    choices.append(final[1][0]);
    choices.append(final[2][0]);
    choices.insert(r1,blank);
    
    return choices;


# In[9]:


alternative_choice = {};
potential_key = {};
best_list ={};
count=1;
simpler_sentences = sent_tokenize(text);

for s in simpler_sentences:
    
    punctuations = '''!()-[]{};:'"“”\,<>./?@#$%^&*_~'''
    for x in s.lower(): 
        if x in punctuations: 
            s = s.replace(x,"");
            
    if len(s.split())<3:
        continue;
        
    key_list = pos_tag(s.split());
    entity_set=nlp(s);
    
    potential_key.clear();
    best_list.clear(); 
    
    if check1(key_list) or check2(entity_set):
        
        print("\n");        
        add(entity_set);     # adding all entities to a database-list...
        get_key(key_list);   # generating potential key list...
        
        # Selecting key with minimum occurance...
        
        blank = list(best_list.keys())[list(best_list.values()).index(min(best_list.values()))];
        s = s.replace(blank,"_____");
        
        # Getting list of alternative choices...
        choices = get_alt(blank);

        print("Que.",count," ",s);
        print("      a- ",choices[0]);
        print("      b- ",choices[1]);
        print("      c- ",choices[2]);
        print("      d- ",choices[3]);
        print();
        print("Ans. ",blank);

        count=count+1;
        


# In[ ]:
```
